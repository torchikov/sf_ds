# Работы с выбросами. Медот межквантильного размаха (метод Тьюки)

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
histplot = sns.histplot(data=sber_data, x="full_sq", ax=axes[0])
histplot.set_title("Full Square Data Distribution")
boxplot = sns.boxplot(data=sber_data, x="full_sq", ax=axes[1])
boxplot.set_title("Full Square Boxplot")
```

```python
def outliers_iqr_mod(data, feature):
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
```

```python
outliers, cleaned = outliers_iqr_mod(sber_data, "full_sq")
print(f"Число выбросов при границах = 1.5 - {outliers.shape[0]}")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
histplot = sns.histplot(data=cleaned, x="full_sq", ax=axes[0])
histplot.set_title("Cleaned Full Square Distribution woth edges = IQR +/- 1.5")
boxplot = sns.boxplot(data=cleaned, x="full_sq", ax=axes[1])
boxplot.set_title("Cleaned Full Square Boxplot with edges = IQR +/- 1.5")
```

```python
def outliers_iqr_mod(data, feature, left=1.5, right=1.5, log_scale=False):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * left)
    upper_bound = quartile_3 + (iqr * right)
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
```

```python
outliers, cleaned = outliers_iqr_mod(sber_data, "full_sq", left=1, right=6)
print(f"Резкльтирующее число выббросов = {outliers.shape[0]}")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
histplot = sns.histplot(data=cleaned, x="full_sq", ax=axes[0])
histplot.set_title("Cleaned Full square distribution")
boxplot = sns.boxplot(data=cleaned, x="full_sq", ax=axes[1])
boxplot.set_title("Cleaned Full Square Boxplot")
```

# Работа с выбросами. Метод z-отклонений (метод сигм)

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
histplot = sns.histplot(data=sber_data["mkad_km"], bins=30, ax=axes[0])
histplot.set_title("MKAD Km Distribution")

log_mkad_km = np.log(sber_data["mkad_km"] + 1)
histplot  = sns.histplot(data=log_mkad_km, bins=30, ax=axes[1])
histplot.set_title("Log MKAD Km Distribution")
```

```python
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 3 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned
```

```python
outliers, cleaned = outliers_z_score(sber_data, "mkad_km", log_scale=True)
print(f"Число выбросов по методу z-отклонений: {outliers.shape[0]}")
print(f"Результирующее число записей: {cleaned.shape[0]}")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 4))
log_mkad_km = np.log(sber_data["mkad_km"] + 1)
histplot = sns.histplot(data=log_mkad_km, bins=30, ax=axes)
histplot.axvline(log_mkad_km.mean(), color="k", lw=2)
histplot.axvline(log_mkad_km.mean() + 3 * log_mkad_km.std(), color="k", ls="--", lw=2)
histplot.axvline(log_mkad_km.mean() - 3 * log_mkad_km.std(), color="k", ls="--", lw=2)
histplot.set_title("Log MKAD Km Distribution")
```

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned
```

```python
outliers, cleaned = outliers_z_score_mod(sber_data, "mkad_km", log_scale=True, left=3, right=3.5)
print(f"Число выбросов по методу z-отклонений: {outliers.shape[0]}")
print(f"Результирующее число записей: {cleaned.shape[0]}")
print(outliers['sub_area'].unique())
```

## Поиск и удаление неинформативных признаков

Чтобы считать признак неинформативным, прежде всего нужно задать какой-то определённый порог. Например, часто используют пороги в 0.95 и 0.99. Это означает: признак неинформативен, если в нем 95 % (99 %) одинаковых значений или же 95 % (99 %) данных полностью уникальны.

```python
# Список неинформативных признаков
low_information_col = []

#цикл по всем столбцам
for col in sber_data.columns:
    #наибольшая относительная частота в признаке
    top_freq = sber_data[col].value_counts(normalize=True).max()
    #доля уникальных значений от размера признака
    nunique_ratio = sber_data[col].nunique() / sber_data[col].count()
    # сравниваем наибольшую частоту с порогом
    if top_freq > 0.95:
        low_information_col.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
    if nunique_ratio > 0.95:
        low_information_col.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
```

```python
information_sber_data = sber_data.drop(low_information_col, axis=1)
print(f'Результирующее число признаков: {information_sber_data.shape[1]}')
```
