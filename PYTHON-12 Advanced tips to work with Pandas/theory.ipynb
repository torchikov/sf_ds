```python
import os 
def concat_user_files(path):
    files = os.listdir(path)
    files.sort()
    result_df = None
    for filename in files:
        full_path = path + "/" + filename
        current_df = pd.read_csv(full_path, sep=",")

        if result_df is None:
            result_df = current_df
        else:
             result_df = pd.concat([result_df, current_df])
        
       
    result_df = result_df.drop_duplicates(ignore_index=True)
    return result_df
    
concat_user_files("data/movies_data")
    
```

```python
import re
def get_year_release(arg):
    candidates = re.findall(r"\(\d{4}\)",arg)
    if len(candidates) > 0:
        year = candidates[0].replace("(", "").replace(")", "")
        return int(year)
    else:
        return None
```

```python
merged["year_release"] = merged["title"].apply(get_year_release)
merged.info()
```

```python
merged[merged["year_release"] == 1999].groupby(by="title")["rating"].mean().sort_values()
```

```python
merged[merged["year_release"] == 2010].groupby(by="genres")["rating"].mean().sort_values()
```

```python
merged.groupby(by="userId")["rating"].agg(["count", "mean"]).sort_values(by=["count", "mean"], ascending=[True, False])
```

```python
merged[merged["year_release"] == 2018].groupby(by="genres")["rating"].agg(["mean", "count"]).sort_values(by=["mean", "count"], ascending=[False, True])
```
